## Representing a relation using Venn Diagrams  

```python
A = set(range(6))
B = set(range(6))
R = {(a,b) for a in A for b in B if (a-b) % 3 == 0} 
draw_relation_as_venn_diagram(A, B, R)

print(R)
```

## Representing a relation using diagraphs

```python
A = set(range(6))
R = {(a,b) for a in A for b in A if (a-b) % 2 == 0} 

draw_relation_on_set_as_digraph(A, R, layout='circular')
```

Or, instead of nodes drawn in a circle, we can draw nodes as if the edges are springs.  This can make is easier to see if the relation is reflexive, symmetric, or transitive.

Now, I will be first to admit that the resulting diagraphs can look messy, but they can still be useful for visualising the properties of a relation. The placement is semi-random, so you may get different layouts each time you draw the diagraph.

```python
draw_relation_on_set_as_digraph(A, R, layout='spring')
```
